```python
import cv2
import matplotlib.cm as cm
import matplotlib.pyplot as plt

dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_6X6_250)

marker_ids = [0, 1, 2, 3]
marker_size = 200

# Show four ArUco markers as a 2x2 grid
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

markers = []
for i, marker_id in enumerate(marker_ids):
    ri, ci = i // 2, i % 2
    marker_image = cv2.aruco.generateImageMarker(dictionary, marker_id, marker_size)
    markers.append(marker_image)
    axs[ri, ci].imshow(marker_image, cmap=cm.gray)
    axs[ri, ci].axis('off')

#plt.imshow(marker_image, cmap=cm.gray)
#plt.axis('off')
plt.show()
```

```python
for i, marker in enumerate(markers):
    plt.imsave(f'aruco_{i}.png', marker, cmap=cm.gray)
```
